```python
#Importing Libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing Dataframe on Jupyter and Reading the same
data = pd.read_csv("Placement_Data_Full_Class.csv") 
```

```python
#Detection of Gender-based Salary Distribution and Outliers using Box Plot Summary 
sns.boxplot(data=data, x="gender", y="salary", palette="Set2")
plt.title("Box Plot for Gender vs Salary")
plt.show()
```

```python
#Detection of Workex-based Salary Distribution and Outliers using Box Plot Summary 
sns.boxplot(data=data, x="workex", y="salary", palette="Set1")
plt.title("Box Plot for Workex vs Salary")
plt.show()
```

```python
#Histogram for Salary Distribution 
plt.figure(figsize=(15,8))
sns.distplot(data["salary"])
plt.title("Distribution of Salary")
plt.show()
```

Inference from above Distribution Curve
Salary distribution is rightly skewed
There are less no. of students with high salary. Majority of them are concentrated between 200000 to 300000

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.countplot(data=data,x="specialisation")
plt.title("Sepecialisation wise Count")
plt.subplot(1,2,2)
sns.countplot(data=data,x="status",palette="Set2")
plt.title("Status wise Count")
plt.show()
```

```python
#Data Validation
data.head()
```

```python
#Scatter Plot and Line Graph between Salary and MBA_P to show a correlation
x=data["mba_p"]
y=data["salary"]
sns.regplot(x = x, y = y,scatter_kws = {"color": "black", "alpha": 0.5}, line_kws = {"color": "red"})
plt.title("Distribution of Salary")
plt.show()
```

### Inferences of above Scatter Plot: The above plot shows that data is positively (moderate) correlated, Increase in MBA_P results in slight increase in Salary

```python
#HeatMap of Triangle Correlation
data_numerical = data._get_numeric_data()  # retrieving only numerical data
data_numerical = data_numerical.drop(columns=data_numerical.columns[0]) # drop the sl_no column
mask = np.triu(np.ones_like(data_numerical.corr(), dtype=np.bool_))
heatmap = sns.heatmap(data_numerical.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16)# heatmap from correlation matrix among numerical variables
plt.show()
```

# Inference from above Traingular Heatmap
1. Salary is more correlated with etest_p and mba_p. Gaining more percentage in Entrance and MBA Exams will results in getting good salary
2. mba_p is more correlated with degree_p
3. hsc_p, ssc_p and degree_p are more or less similarly correlated with etest_p
4. degree_p is more correlated with ssc_p compared to hsc_p
5. hsc_p is correlated with ssc_p
