```python
import seaborn as sns
import matplotlib.pyplot as plt # visualizing data
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Count of Each Gender')
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

```python
X = df.groupby(['Gender'])['Amount'].sum().reset_index()
print(X)
plt.title('Total Amount Spent by Gender')
sns.barplot(x='Gender',y='Amount', data= X)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

```python
plt.figure(figsize=(10, 6))
s=['0-17','18-25','26-35','36-45','46-50','51-55','55+']
sns.countplot(x='Age Group', hue='Gender', data=df, order=s)
plt.title('Count of Gender Age Group-wise')
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

```python
Y = df.groupby(['Age Group'])['Amount'].sum().reset_index()
print(Y)
plt.title('Total Amount Spent by different Age Group')
sns.barplot(x='Age Group',y='Amount', data= Y)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

```python
plt.figure(figsize=(22, 6))
Z = df.groupby(['State'])['Orders'].sum().reset_index()
print(Z)
plt.title('Total Orders by State')
sns.barplot(x='State',y='Orders', data= Z)
```

```python
plt.figure(figsize=(22, 6))
A = df.groupby(['State'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
print(A)
plt.title('Total Amount Spent on Orders by State')
sns.barplot(x='State',y='Amount', data= A)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


```python
plt.figure(figsize=(5, 6))
sns.countplot(x='Marital_Status', data=df)
plt.title('Count by Marital Status')
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
s=['0-17','18-25','26-35','36-45','46-50','51-55','55+']
sns.countplot(x='Age Group', hue='Marital_Status', data=df, order=s)
plt.title('Count by Marital Status Age Group-wise')
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

```python
plt.figure(figsize=(5, 7))
B = df.groupby(['Marital_Status'])['Amount'].sum().reset_index()
print(B)
plt.title('Total Amount Spent Marital Status wise')
sns.barplot(x='Marital_Status',y='Amount', data= B)
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

```python
C = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().reset_index()
print(C)
plt.figure(figsize=(8, 6))
sns.barplot(data = C, x = 'Marital_Status', y= 'Amount', hue='Gender')
plt.title('Amount Spent on Orders by Marital Status Age Group-wise')
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()



```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

```python
plt.figure(figsize=(20, 6))
sns.countplot(x='Occupation', data=df, order=df['Occupation'].value_counts().index)
plt.title('Count by Occupation')
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

```python
plt.figure(figsize=(20, 7))
D = df.groupby(['Occupation'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
print(D)
plt.title('Total Amount Spent Occupation wise')
sns.barplot(x='Occupation',y='Amount', data= D)
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

```python
plt.figure(figsize=(35, 10))
sns.countplot(x='Product_Category', data=df, order=df['Product_Category'].value_counts().index)
plt.title('Count by Product_Category')
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```
